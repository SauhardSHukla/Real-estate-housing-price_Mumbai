```python
#import of the Lib.
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Import the file for the performing all the tasks
df = pd.read_csv('C:/Users/5530/OneDrive/Desktop/Real_State_Housing_Mumbai/Creation & Handling of CSV files/CSV_Files/simple_sampling.csv')
# df.head(2)
```

##### 1.One-sample Testing

--------------

***Test if the average price of the properties in the sampled dataset is equal to ₹1.2 Crore (12,000,000).***

```python
# A competitor’s average price (assumed for the query)
mu = 130     #(price in lakh of the property)

# Perform one-sample t-test on the 'price' column
t_stat, p_value = stats.ttest_1samp(df['price_lakhs'],popmean=mu)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.5f}") 
```

***Reject VS Success***

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: Mean price is significantly different from ₹1.2 Cr")
else:
    print("Fail to reject the null hypothesis: No significant difference from ₹1.2 Cr")
```

***Graphical Viusalization***

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price_lakhs'], bins=30, kde=True, color='skyblue')

#Mean price of the Dataframe
mean = df['price_lakhs'].mean()

# Mark the sample mean and hypothesized mean
plt.axvline(mean, color='green', linestyle='--', label=f'Sample Mean = ₹{mean:.2f} lakhs')
plt.axvline(130, color='red', linestyle='--', label='Hypothesized Mean = ₹130 lakhs')

plt.title("Distribution of Property Prices")
plt.xlabel("Price (in Lakhs)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

***Context Recap***

You took a random sample of property prices in Mumbai and tested whether the average price is equal to ₹130 lakhs (₹1.3 crores) — a business benchmark, expectation, or past-year average.


#### Business Insights:-

***Loan Policy***

***Testing Average Size of Properties in Mumbai***


***Business Question:***
 Is the average size of properties in Mumbai greater than 1000 sq.ft?

```python
# A competitor’s average price (assumed for the query)
mu = 750    #(sqft )

# Perform one-sample t-test on the 'price' column
t_stat, p_value = stats.ttest_1samp(df['size_sqft'],popmean=mu)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.5f}") 
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: Mean price is significantly different from 1000 sqft")
else:
    print("Fail to reject the null hypothesis: No significant difference from 1000 sqft")
```

***Graphical Viusalization***

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['size_sqft'], bins=30, kde=True, color='skyblue')

#Mean price of the Dataframe
mean = df['size_sqft'].mean()

# Mark the sample mean and hypothesized mean
plt.axvline(mean, color='green', linestyle='--', label=f'Sample Mean = {mean:.2f} sqft')
plt.axvline(1000, color='red', linestyle='--', label='Hypothesized Mean = 1000 sqft')

plt.title("Distribution of Property Prices")
plt.xlabel("Price (in Lakhs)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

###  Market Area Size is Smaller than Assumed (855 sqft < 1000 sqft)

- **Valuation Correction**  
  Update area assumptions to ~855 sqft, not 1000 sqft.

- **Loan Adjustments**  
  Loan limits and EMIs should reflect **smaller property sizes**.

- **Design Strategy**  
  Developers should consider **compact housing models**.

- **Marketing Strategy**  
  Target **budget-conscious or first-time buyers** who prefer smaller units.

- **Pricing Packages**  
  Consider launching **affordable housing** or **studio apartments**.

- **Buyer Advisory**  
  Prepare buyers for **smaller living spaces** — reset expectations.

- **Investor Strategy**  
  Investors can focus on **high-volume, smaller unit sales** for better ROI.


#### In SHORT

***Testing the Avg BHK size of the property in Mumbai***

```python
bhk_data = df['bhk'].dropna()

# Perform one-sample t-test: Is mean(BHK) different from 2?
t_stat, p_value = stats.ttest_1samp(bhk_data, popmean=2)

print("Mean BHK:", bhk_data.mean())
print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

***Reject VS Success***

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: Mean price is significantly different from 2 BHK")
else:
    print("Fail to reject the null hypothesis: No significant difference from 2 BHK")
```

***Graphical Viusalization***

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['bhk'], bins=30, color='skyblue')

#Mean price of the Dataframe
mean = df['bhk'].mean()

# Mark the sample mean and hypothesized mean
plt.axvline(mean, color='green', linestyle='--', label=f'Sample Mean = {mean:.2f} ')
plt.axvline(2, color='red', linestyle='--', label='Hypothesized Mean = 2 BHK')

plt.title("Distribution of Property Prices")
plt.xlabel("BHK")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```


###  Business Insights from BHK (One-Sample T-Test)
**Assumed Mean = 2**, **Actual Mean = 2.10**


- **Property Valuation**  
  Slight increase in average BHK (~2.10) justifies **slightly higher price ranges** for listings.

- **Target Audience Adjustment**  
  Indicates a demand shift toward **larger households** (small families or shared tenants), not just single buyers.

- **Marketing Strategy**  
  Focus on promoting **spacious floor plans** or **extra rooms**, especially for 2.5BHK and 3BHK units.

- **Construction & Development**  
  Builders can plan more **2.5BHK or convertible units**, reflecting current buying behavior.

- **Loan and Financial Advisory**  
  Suggest **increased loan limits** or financing options to accommodate slightly higher average room requirements.

***Testing the montly rent of the various property across Mumbai***

```python

assumed_rent = 40000 

# Step 2: Calculate sample mean
mean_rent = df['rent_monthly'].mean()
print(f"Sample Mean Rent: ₹{mean_rent:.2f}")

# Step 3: Perform one-sample t-test
t_stat, p_value =stats.ttest_1samp(df['rent_monthly'].dropna(), assumed_rent)
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.40f}")
```

***Reject VS Success***

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: Mean price is significantly different from Montly rent ")
else:
    print("Fail to reject the null hypothesis: No significant difference from Montly rent")
```

***Graphical Visualization***

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['rent_monthly'], bins=30, color='skyblue')

#Mean price of the Dataframe
mean = df['rent_monthly'].mean()

# Mark the sample mean and hypothesized mean
plt.axvline(mean, color='green', linestyle='--', label=f'Sample Mean = {mean:.2f} ')
plt.axvline(40000, color='red', linestyle='--', label='Hypothesized Mean = 2 BHK')

plt.title("Distribution of Property Prices")
plt.xlabel("rent_monthly")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

####  One-Sample T-Test Insight on Monthly Rent

-  **Assumed average rent**: ₹40,000  
-  **Observed mean rent**: ₹44,598  
-  **p-value**: 0.0021  
-  **T-statistic**: 3.11

#### Interpretation:
The p-value is less than 0.05, indicating a statistically significant difference between the assumed and actual average rent.

####  Business Insight:
The average rent in the current listings is significantly **higher** than the market expectation (₹40,000). This suggests:
- Tenants might perceive the area as **premium** or in **high demand**.
- Landlords can **justify current rent levels**.
- Property investors may consider this area for **rental yield opportunities**.
