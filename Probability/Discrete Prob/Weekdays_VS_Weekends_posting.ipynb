#### We want to determine whether more properties are posted on weekends (Saturday or Sunday), using the posted_on column and applying Bernoulli distribution logic

***Import the Lib***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***Load the CSV File***

```python
df = pd.read_csv('C:/Users/5530/OneDrive/Desktop/Real_State_Housing_Mumbai/Creation & Handling of CSV files/CSV_Files/simple_sampling.csv')
```

***Handling the Datetime columns in the Database***

```python
df['posted_on'] = pd.to_datetime(df['posted_on'])

df['is_weekend'] = (df['posted_on'].dt.dayofweek>=5).astype(int)
```

***Calculate probabilities***

```python
total = len(df)
weekend = df['is_weekend'].sum()
weekday = total - weekend
p = weekend / total

print(f"Sample Size: {total}")
print(f"Weekend Listings: {weekend} ({p:.2%})")
print(f"Weekday Listings: {weekday} ({1 - p:.2%})")
```

***Visual***

```python
sns.histplot(data=df, x ='is_weekend')
plt.xticks([0,1],['Weekday','Weekend'])
plt.title('Sampled Data: Weekday vs Weekend Listings')
plt.xlabel('Day Type')
plt.ylabel('Number of Listings')
plt.tight_layout()
plt.grid(True)
plt.show()
```

***Only 26% of listings are posted on weekends, suggesting:***

Most agents or users prefer weekdays (possibly business hours).

You could recommend higher visibility or promotions on weekdays.
